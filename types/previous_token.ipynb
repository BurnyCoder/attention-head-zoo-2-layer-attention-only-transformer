# Previous Token Head

Attends to the immediately preceding token

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import torch as t
from IPython.display import Markdown, display
from shared import (
    load_model, run_and_cache, get_attention_pattern,
    show_head_pattern, show_attention_tables,
    compute_all_type_metrics, HEAD_TYPES, ACTIVITY_LABELS,
    get_head_types, TEXT,
)
```

```python
model = load_model()
str_tokens, logits, cache = run_and_cache(model)
```

## Previous Token Head — All 24 Heads

```python
tm = compute_all_type_metrics(cache, str_tokens)
is_measurable = ("previous_token", 0, 0) in tm
if is_measurable:
    results = sorted(
        [((l, h), tm[("previous_token", l, h)]) for l in range(2) for h in range(12)],
        key=lambda x: x[1], reverse=True,
    )
    lines = ["| Head | Previous Token Head % |", "|------|-------|"]  
    for (l, h), pct in results:
        lines.append(f"| L{l}H{h} | {pct:.2f}% |")
    display(Markdown("\n".join(lines)))
else:
    print("No programmatic metric available for this type.")
```

```python
pct = tm[("previous_token", 0, 7)]
level = 'full' if pct >= 90 else 'fullish' if pct >= 60 else 'half' if pct >= 40 else 'partial' if pct >= 10 else 'almost none' if pct >= 0.1 else '-'
display(Markdown(f"---\n## L0H7 — {pct:.2f}% ({level})\n\nAttends only to previous token, one to semantically salient (scaled up)"))
```

```python
show_head_pattern(str_tokens, cache, layer=0, head=7)
```

```python
attention = get_attention_pattern(cache, layer=0, head=7)
show_attention_tables(str_tokens, attention, top_k=25)
```
