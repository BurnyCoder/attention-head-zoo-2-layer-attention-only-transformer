## Related Previous Token — All 24 Heads

```python
tm = compute_all_type_metrics(cache, str_tokens)
is_measurable = ("related_previous_token", 0, 0) in tm
if is_measurable:
    results = sorted(
        [((l, h), tm[("related_previous_token", l, h)]) for l in range(2) for h in range(12)],
        key=lambda x: x[1], reverse=True,
    )
    lines = ["| Head | Related Previous Token % |", "|------|-------|"]  
    for (l, h), pct in results:
        lines.append(f"| L{l}H{h} | {pct:.2f}% |")
    display(Markdown("\n".join(lines)))
else:
    print("No programmatic metric available for this type.")
```
