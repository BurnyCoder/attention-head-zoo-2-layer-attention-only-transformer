# Few Previous Tokens Head

Attends to a small window of preceding tokens

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import torch as t
from IPython.display import Markdown, display
from shared import (
    load_model, run_and_cache, get_attention_pattern,
    show_head_pattern, show_attention_tables,
    compute_all_type_metrics, HEAD_TYPES, TYPE_ENTROPY_KEYS,
    ACTIVITY_LABELS, get_head_types, TEXT,
)
```

```python
model = load_model()
str_tokens, logits, cache = run_and_cache(model)
```

## Few Previous Tokens Head — All 24 Heads

```python
tm = compute_all_type_metrics(cache, str_tokens)
ent_key = TYPE_ENTROPY_KEYS.get("few_previous_tokens")
is_measurable = ("few_previous_tokens", 0, 0) in tm
if is_measurable:
    results = sorted(
        [((l, h), tm[("few_previous_tokens", l, h)]) for l in range(2) for h in range(12)],
        key=lambda x: x[1], reverse=True,
    )
    has_ent = ent_key and (ent_key, 0, 0) in tm
    if has_ent:
        lines = ["| Head | Few Previous Tokens Head % | Entropy % |", "|------|-------|-------|"]  
        for (l, h), pct in results:
            ent = tm[(ent_key, l, h)]
            lines.append(f"| L{l}H{h} | {pct:.2f}% | {ent:.2f}% |")
    else:
        lines = ["| Head | Few Previous Tokens Head % |", "|------|-------|"]  
        for (l, h), pct in results:
            lines.append(f"| L{l}H{h} | {pct:.2f}% |")
    display(Markdown("\n".join(lines)))
else:
    print("No programmatic metric available for this type.")
```

```python
pct = tm[("few_previous_tokens", 0, 4)]
ent_str = f" | ent {tm[(ent_key, 0, 4)]:.2f}%" if ent_key and (ent_key, 0, 4) in tm else ""
level = 'full' if pct >= 90 else 'fullish' if pct >= 60 else 'half' if pct >= 40 else 'partial' if pct >= 10 else 'almost none' if pct >= 0.1 else '-'
display(Markdown(f"---\n## L0H4 — {pct:.2f}% ({level}){ent_str}\n\nAttends to few previous tokens"))
```
