## Few Previous Tokens Attention % Across All 24 Heads

Mean attention to the 5 preceding tokens (excluding self). Sorted by raw % descending.

```python
show_few_prev_tokens_pcts(cache, k=5)
```

```python
attn = get_attention_pattern(cache, layer=0, head=4)
n = attn.shape[0]
pct = sum(attn[i, max(0, i-5):i].sum().item() for i in range(n)) / n * 100
level = 'full' if pct >= 90 else 'fullish' if pct >= 60 else 'half' if pct >= 40 else 'partial' if pct >= 10 else 'almost none' if pct >= 0.1 else '-'
display(Markdown(f"---\n## L0H4 — {pct:.2f}% ({level})\n\nAttends to few previous tokens"))
```
